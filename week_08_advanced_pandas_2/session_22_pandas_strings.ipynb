```python
# import titanic
df = pd.read_csv('titanic.csv')
df['Name']
```

```python
# Common Functions
# lower/upper/capitalize/title
df['Name'].str.upper()
df['Name'].str.capitalize() # first letter of first word is capital and rest all small
df['Name'].str.title() # first letter of all words capital and rest all small
# len
df['Name'][df['Name'].str.len() == 82].values[0]
# strip
"                   nitish                              ".strip()
df['Name'].str.strip() # removes all leading and trailing spaces
```

```python
df['Name'].str.len().max()
```

```python
df['Name'][df['Name'].str.len() == 82]
```

```python
df['Name'][df['Name'].str.len() == 82].values[0] # finding the longest name
```

```python
# split -> get
df['lastname'] = df['Name'].str.split(',').str.get(0) # this line splits at , first and then gets the first element
# basically Braund, Mr. Owen Harris will be split at , and Braund will be saved to lastname column
df.head()
```

```python
df[['title','firstname']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand=True)
# we are splitting the first name on the basis of first space and then we are not splitting
# we want it to be like Mr. Owen Harris -> we dont want to split again as owen harris , n=1 means only one split
# expand=True changes it to dataframe
df.head()

```

```python
df['title'].value_counts()
```

```python
# replace
df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
```

```python
df['title'].value_counts()
```

```python
# filtering
# startswith/endswith
df[df['firstname'].str.startswith('A')]
```

```python
df[df['firstname'].str.endswith('A')]
```

```python
# isdigit/isalpha...
df[df['firstname'].str.isdigit()]
```

```python
df[df['firstname'].str.isalpha()] #  
```

```python
# applying regex
# contains
# search john -> both case
df[df['firstname'].str.contains('john',case=False)]
# find lastnames with start and end char vowel
df[df['lastname'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```
