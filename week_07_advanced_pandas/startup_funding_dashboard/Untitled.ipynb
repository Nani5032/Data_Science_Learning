```python
import pandas as pd
df = pd.read_csv('startup_funding.csv')
```

```python
df.head()
```

```python
df['Investors Name'] = df['Investors Name'].fillna('Undisclosed')
# filling the empty investors name with Undisclosed
```

```python
df
```

```python
sorted(df['Investors Name'].unique().tolist())
# here we are checking the unique investors
```

```python
df.drop(columns=['Remarks'], inplace=True)
```

```python
df.head()
```

```python
df.set_index('Sr No', inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={
    'Date dd/mm/yyyy':'date',
    'Startup Name':'startup',
    'Industry Vertical':'vertical',
    'SubVertical':'subvertical',
    'City  Location':'city',
    'Investors Name':'investors',
    'InvestmentnType':'round',
    'Amount in USD':'amount'    
    
},inplace=True)
```

```python
df.head()
```

```python
# now we are converting amount to inr in crores and also the amount column is in string format and there are some missing values
# to handle this first we are putting string 0 like this '0'
df['amount'] = df['amount'].fillna('0')
```

```python
df.info()
```

```python
df['amount'] = df['amount'].str.replace(',','') # instring we have 50,000 we need to remove , else we cannot convert string to integer
df['amount'] = df['amount'].str.replace('undisclosed','0') # replacing undisclosed with 0
df['amount'] = df['amount'].str.replace('unknown','0') # replacing unknown with 0
df['amount'] = df['amount'].str.replace('Undisclosed','0') #  replacing Undisclosed with 0
```

```python
df['amount'].str.isdigit().value_counts() # there are some characters like 1428000+ which we cannot change
# for that we are checking if the number is a digit or not and then considering them
```

```python
df = df[df['amount'].str.isdigit()]
```

```python
df['amount'] = df['amount'].astype('float')
```

```python
df.head()
```

```python
df.info()
```

```python
def to_inr(dollar):
    inr = dollar * 82.5
    return inr/10000000
```

```python
df['amount'] = df['amount'].apply(to_inr)
```

```python
df.head()
```

```python
df['date'] = df['date'].str.replace('05/072018','05/07/2018')
```

```python
df['date'] = pd.to_datetime(df['date'],errors='coerce') # coerce is we are ignoring the date with issue
# we will look into coerce thing later
```

```python
df = df.dropna(subset=['date','startup','vertical','city','investors','round','amount'])
```

```python
df.info()
```

```python
df.to_csv('startup_cleaned.csv',index=False)
```
